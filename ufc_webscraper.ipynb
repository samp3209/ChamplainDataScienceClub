```python
# Read URLs from the CSV file
urls_df = pd.read_csv("UFC - URLS (1).csv")
url_list = urls_df['URLs'].tolist()

# Initialize lists to store the scraped data
all_Aname = []
all_Bname = []
all_odds = []
all_win = []
all_round = []

# Loop through the list of URLs
for url in url_list:
    response = requests.get(url)
    if response.status_code == 200:
        html_content = response.content
        soup = BeautifulSoup(html_content, 'html.parser')

        # Extract data from the current URL
        Aname = [x.text.strip() for x in soup.find_all("div", {"class": "c-listing-fight__corner-name c-listing-fight__corner-name--red"})]
        Bname = [x.text.strip() for x in soup.find_all("div", {"class": "c-listing-fight__corner-name c-listing-fight__corner-name--blue"})]
        odds = [x.text.strip() for x in soup.find_all("span", {"class": "c-listing-fight__odds-amount"})]
        win = [x.text.strip() for x in soup.find_all("div", {"class": "c-listing-fight__result-text method"})]
        round = [x.text.strip() for x in soup.find_all("div", {"class": "c-listing-fight__result-text round"})]

        # Append the data to the respective lists
        all_Aname.extend(Aname)
        all_Bname.extend(Bname)
        all_odds.extend(odds)
        all_win.extend(win)
        all_round.extend(round)

rounds = all_round[::2]
methods = all_win[::2]
oddsA = all_odds[::2]
oddsB = all_odds[1::2]
# Create a DataFrame to store the scraped data


```

```python
data = {
    'Aname': all_Aname,
    'Bname': all_Bname,
    'oddsA': oddsA,
    'oddsB': oddsB,
    'method': methods,
    'endround': rounds
}

scraped_data_df = pd.DataFrame(data)
```

```python
scraped_data_df.to_csv('modernUFC.csv')
```
